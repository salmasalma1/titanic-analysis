```python

# 1. Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 2. Load Data
train_url = "https://raw.githubusercontent.com/agconti/kaggle-titanic/master/data/train.csv"
test_url = "https://raw.githubusercontent.com/agconti/kaggle-titanic/master/data/test.csv"

train = pd.read_csv(train_url)
test = pd.read_csv(test_url)

# 3. Quick Preview
print("=== Training Data (First 5 Rows) ===")
print(train.head())

# 4. Data Exploration
print("\n=== Data Summary ===")
print(f"Training Shape: {train.shape}")
print(f"Test Shape: {test.shape}")

print("\n=== Missing Values ===")
print(train.isnull().sum())

# 5. Data Cleaning (Drop unnecessary columns)
train_clean = train.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1)

# Fill missing values
train_clean['Age'].fillna(train_clean['Age'].median(), inplace=True)
train_clean['Embarked'].fillna(train_clean['Embarked'].mode()[0], inplace=True)

# Convert gender to binary (male=1, female=0)
train_clean['Sex'] = train_clean['Sex'].map({'male': 1, 'female': 0})

# 6. Data Visualization
plt.figure(figsize=(15, 10))

# Plot 1: Survival Count
plt.subplot(2, 2, 1)
sns.countplot(x='Survived', data=train_clean)
plt.title("Survival Distribution (0=Died, 1=Survived)")

# Plot 2: Survival by Gender
plt.subplot(2, 2, 2)
sns.countplot(x='Survived', hue='Sex', data=train_clean)
plt.title("Survival by Gender (1=Male, 0=Female)")

# Plot 3: Survival Rate by Passenger Class
plt.subplot(2, 2, 3)
sns.barplot(x='Pclass', y='Survived', data=train_clean)
plt.title("Survival Rate by Ticket Class")

# Plot 4: Age Distribution
plt.subplot(2, 2, 4)
sns.histplot(train_clean['Age'], bins=20, kde=True)
plt.title("Passenger Age Distribution")

plt.tight_layout()
plt.show()

```
